```python
import pandas as pd

# Overall data
train_data = pd.read_pickle(open("tweets-data/train.pkl", "rb"))

pd.set_option('display.max_colwidth', None)
train_data
```

```python
def is_noise(df):
    result = []
    for row in range(len(df)):
        line = df.loc[row, 'text']

        if '_TWITTER-ENTITY_' in line or line[0] == '\\':
            result.append(True)
        else:
            result.append(False)
    return result
```

### Split noise data

```python
noise = train_data.loc[is_noise(train_data)]
noise
```

```python
AAE_noise = noise.loc[noise['Demographic'] == 'AAE']
print("Number of noise in AAE:", len(AAE_noise))
positive_ratio = AAE_noise.loc[AAE_noise['Sentiment'] == 'positive']
print("Positive ratio in AAE: {}".format(round(len(positive_ratio) / len(AAE_noise), 3)))
```

```python
SAE_noise = noise.loc[noise['Demographic'] == 'SAE']
print("Number of noise in SAE:", len(SAE_noise))
positive_ratio = SAE_noise.loc[SAE_noise['Sentiment'] == 'positive']
print("Positive ratio in SAE: {}".format(round(len(positive_ratio) / len(SAE_noise), 3)))
```

### Filtered clean data

```python
clean = train_data.loc[[not elem for elem in is_noise(train_data)]]
clean
```
