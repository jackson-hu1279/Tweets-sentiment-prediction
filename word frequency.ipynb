### Most frequent words

```python
def frequent_words(data, file_path, word_number):
    agg_ifidf = None
    for row in range(len(data)):
        line = data.loc[row, 'TFIDF']
        if agg_ifidf is None:
            agg_ifidf = line
        else:
            agg_ifidf = np.add(agg_ifidf, line)

    indices = sorted(range(len(agg_ifidf)), key = lambda sub: agg_ifidf[sub])[-word_number:]
    indices.reverse()
    print("Index of most frequent words:", indices)

    all_words = pd.read_csv(file_path, sep='\t')
    most_frequent = {}
    for index in indices:
        most_frequent[all_words.loc[index, 'w']] = agg_ifidf[index]
    
    # print(most_frequent)
    print("Top {} words:".format(word_number))
    for word in most_frequent.items():
        print(word)
```

```python
frequent_words(AAE_train_data, "tfidf/tfidf_word_idx.txt", 10)
```

```python
frequent_words(SAE_train_data, "tfidf/tfidf_word_idx.txt", 10)
```

### Analyse some feature words

```python
def word_statistics(data, word):
    total_count = AAE_count = SAE_count = 0
    total_record = len(data)

    for row in range(total_record):
        line = data.loc[row, 'text']
        if word in line.split():
            total_count += 1
            if data.loc[row, 'Demographic'] == 'AAE':
                AAE_count += 1
            elif data.loc[row, 'Demographic'] == 'SAE':
                SAE_count += 1
    
    if total_count == 0:
        print("Doesn't exist")
        return

    print("Total count of {}: {}".format(word, total_count))
    print("AAE count: {}, ratio: {}".format(AAE_count, round(AAE_count/total_count, 3)))
    print("SAE count: {}, ratio: {}".format(SAE_count, round(SAE_count/total_count, 3)))
```

```python
word_statistics(raw_data, 'going')
print()
word_statistics(raw_data, 'goin')
```

```python
word_statistics(raw_data, 'relax')
print()
word_statistics(raw_data, 'chill')
```
